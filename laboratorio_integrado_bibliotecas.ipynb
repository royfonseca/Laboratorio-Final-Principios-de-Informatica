La **transpuesta** de una matriz $A$ (denotada $A^T$) se obtiene intercambiando filas por columnas:

<summary><h3>â• 4. Operaciones con Vectores</h3></summary>

### Suma de Vectores
Se suman componente a componente. Ambos vectores deben tener la misma dimensiÃ³n:

$$\vec{v} + \vec{w} = [v_1 + w_1, v_2 + w_2, ..., v_n + w_n]$$

**Ejemplo:**
$$[1, 2, 3] + [4, 5, 6] = [5, 7, 9]$$

### MultiplicaciÃ³n por Escalar
Se multiplica cada componente por el escalar $c$:

$$c \cdot \vec{v} = [c \cdot v_1, c \cdot v_2, ..., c \cdot v_n]$$

**Ejemplo:**
$$2.5 \times [1, 2, 3] = [2.5, 5.0, 7.5]$$

### Producto Punto (Dot Product)
Suma de los productos de componentes correspondientes:

$$\vec{v} \cdot \vec{w} = v_1 \cdot w_1 + v_2 \cdot w_2 + ... + v_n \cdot w_n = \sum_{i=1}^{n} v_i \cdot w_i$$

**Ejemplo:**
$$[1, 2, 3] \cdot [4, 5, 6] = 1 \times 4 + 2 \times 5 + 3 \times 6 = 4 + 10 + 18 = 32$$

**âš ï¸ Importante:** Los vectores deben tener la misma dimensiÃ³n.

### Norma (Magnitud)
La longitud o magnitud del vector:

$$\|\vec{v}\| = \sqrt{v_1^2 + v_2^2 + ... + v_n^2} = \sqrt{\vec{v} \cdot \vec{v}}$$

**Ejemplo (Teorema de PitÃ¡goras):**
$$\|[3, 4]\| = \sqrt{3^2 + 4^2} = \sqrt{9 + 16} = \sqrt{25} = 5$$

<summary><h3>ğŸ”¢ 5. Operaciones con Matrices</h3></summary>

### Suma de Matrices
Se suman elemento a elemento. Las matrices deben tener la **misma forma** (mismo nÃºmero de filas y columnas):

$$C[i][j] = A[i][j] + B[i][j]$$

**Ejemplo:**
$$\begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix} + \begin{bmatrix} 5 & 6 \\ 7 & 8 \end{bmatrix} = \begin{bmatrix} 6 & 8 \\ 10 & 12 \end{bmatrix}$$

### MultiplicaciÃ³n por Escalar
Se multiplica cada elemento por el escalar $c$:

$$B[i][j] = c \cdot A[i][j]$$

**Ejemplo:**
$$2 \times \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix} = \begin{bmatrix} 2 & 4 \\ 6 & 8 \end{bmatrix}$$

### MultiplicaciÃ³n Matriz-Vector
Cuando multiplicamos matriz $A_{m \times n}$ por vector $\vec{v}_n$, el resultado es un vector de dimensiÃ³n $m$:

$$\text{resultado}[i] = \sum_{k=1}^{n} A[i][k] \cdot v[k]$$

Es decir, **cada elemento del resultado es el producto punto de una fila de $A$ con $\vec{v}$**.

**Ejemplo:**
$$\begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \end{bmatrix} \times \begin{bmatrix} 1 \\ 2 \\ 3 \end{bmatrix} = \begin{bmatrix} 1 \cdot 1 + 2 \cdot 2 + 3 \cdot 3 \\ 4 \cdot 1 + 5 \cdot 2 + 6 \cdot 3 \end{bmatrix} = \begin{bmatrix} 14 \\ 32 \end{bmatrix}$$

**âš ï¸ Regla de compatibilidad:** El nÃºmero de columnas de $A$ debe ser igual a la dimensiÃ³n de $\vec{v}$.

### MultiplicaciÃ³n Matriz-Matriz
Para multiplicar $A_{m \times n}$ por $B_{n \times p}$, el resultado es $C_{m \times p}$:

$$C[i][j] = \sum_{k=1}^{n} A[i][k] \cdot B[k][j]$$

**InterpretaciÃ³n:** $C[i][j]$ es el producto punto de la **fila $i$ de $A$** con la **columna $j$ de $B$**.

**Ejemplo:**
$$\begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}_{2 \times 2} \times \begin{bmatrix} 5 & 6 \\ 7 & 8 \end{bmatrix}_{2 \times 2} = \begin{bmatrix} 1 \cdot 5 + 2 \cdot 7 & 1 \cdot 6 + 2 \cdot 8 \\ 3 \cdot 5 + 4 \cdot 7 & 3 \cdot 6 + 4 \cdot 8 \end{bmatrix} = \begin{bmatrix} 19 & 22 \\ 43 & 50 \end{bmatrix}_{2 \times 2}$$

**âš ï¸ Regla de compatibilidad:** El nÃºmero de columnas de $A$ debe ser igual al nÃºmero de filas de $B$.

<summary><h3>ğŸ¯ 6. Determinante (â­ BONUS - Avanzado)</h3></summary>

### Â¿QuÃ© es el Determinante?
El **determinante** es un nÃºmero escalar que se puede calcular para cualquier matriz cuadrada. Tiene propiedades geomÃ©tricas importantes:
- Mide el "volumen" de la transformaciÃ³n lineal
- Si det = 0, la matriz no es invertible

NotaciÃ³n: $\det(A)$ o $|A|$

### Matriz 2Ã—2 - FÃ³rmula Directa

$$\det\begin{bmatrix} a & b \\ c & d \end{bmatrix} = ad - bc$$

**Ejemplo:**
$$\det\begin{bmatrix} 4 & 3 \\ 2 & 1 \end{bmatrix} = 4 \times 1 - 3 \times 2 = 4 - 6 = -2$$

### Matriz 3Ã—3 - Regla de Sarrus

$$\det\begin{bmatrix} a & b & c \\ d & e & f \\ g & h & i \end{bmatrix} = aei + bfg + cdh - ceg - afh - bdi$$

**MÃ©todo visual:** Diagonal principal (+) vs diagonal secundaria (-)

### ExpansiÃ³n por Cofactores (MÃ©todo General)
Para matrices $n \times n$ donde $n \geq 3$:

$$\det(A) = \sum_{j=1}^{n} (-1)^{i+j} \cdot A[i][j] \cdot \det(M_{ij})$$

donde $M_{ij}$ es la **submatriz** que resulta de eliminar la fila $i$ y columna $j$.

**Pasos:**
1. Elegir una fila o columna (tÃ­picamente la primera fila)
2. Para cada elemento en esa fila:
   - Calcular el cofactor: $(-1)^{fila+columna} \times elemento$
   - Calcular el determinante de la submatriz (recursivamente)
   - Multiplicar cofactor por determinante de submatriz
3. Sumar todos los resultados

**Ejemplo 3Ã—3:**
$$\det\begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{bmatrix} = 1 \cdot \det\begin{bmatrix} 5 & 6 \\ 8 & 9 \end{bmatrix} - 2 \cdot \det\begin{bmatrix} 4 & 6 \\ 7 & 9 \end{bmatrix} + 3 \cdot \det\begin{bmatrix} 4 & 5 \\ 7 & 8 \end{bmatrix}$$

### Propiedades Ãštiles
1. $\det(I) = 1$ (determinante de identidad)
2. $\det(A^T) = \det(A)$
3. Si una fila/columna es todo ceros, $\det(A) = 0$
4. Intercambiar dos filas cambia el signo: $\det = -\det$
